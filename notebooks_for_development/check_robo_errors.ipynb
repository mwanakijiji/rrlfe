```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/test_ews_20200525.csv")
```

```python
df.keys()
```

```python
plt.scatter(df["balmer"],df["K"])
plt.show()
```

```python
bad1 = df.where(df["original_spec_file_name"] == "600025p00.smo").dropna()
bad2 = df.where(df["original_spec_file_name"] == "600030p02.smo").dropna()
bad3 = df.where(df["original_spec_file_name"] == "625020p00.smo").dropna()
bad4 = df.where(df["original_spec_file_name"] == "625025p02.smo").dropna()
bad5 = df.where(df["original_spec_file_name"] == "575020p02.smo").dropna()
bad6 = df.where(df["original_spec_file_name"] == "600020p02.smo").dropna()
```

```python
plt.scatter(df["balmer"],df["err_balmer"])

plt.scatter(bad1["balmer"],bad1["err_balmer"], label="600025p00.smo")
plt.scatter(bad2["balmer"],bad2["err_balmer"], label="600030p02.smo")
plt.scatter(bad3["balmer"],bad3["err_balmer"], label="625020p00.smo")
plt.scatter(bad4["balmer"],bad4["err_balmer"], label="625025p02.smo")
plt.scatter(bad5["balmer"],bad5["err_balmer"], label="575020p02.smo")
plt.scatter(bad6["balmer"],bad6["err_balmer"], label="600020p02.smo")

plt.legend()
plt.xlabel("Robo Balmer EW")
plt.ylabel("Robo error in Balmer EW")
plt.savefig("new_balmer_comparisons.pdf")
```

```python
plt.scatter(df["K"],df["err_K"])

plt.scatter(bad1["K"],bad1["err_K"], label="600025p00.smo")
plt.scatter(bad2["K"],bad2["err_K"], label="600030p02.smo")
plt.scatter(bad3["K"],bad3["err_K"], label="625020p00.smo")
plt.scatter(bad4["K"],bad4["err_K"], label="625025p02.smo")
plt.scatter(bad5["K"],bad5["err_K"], label="575020p02.smo")
plt.scatter(bad6["K"],bad6["err_K"], label="600020p02.smo")

plt.legend()
plt.xlabel("Robo CaIIK EW")
plt.ylabel("Robo error in CaIIK EW")
plt.savefig("new_caiik_comparisons.pdf")
```
