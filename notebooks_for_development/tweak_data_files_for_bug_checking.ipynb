```python
# This just reads in *.dat spectral files and calculates errors as sqrt(flux), so that the 
# pipeline can use them

# Created 2018 Aug 28 by E.S.
```

```python
import pandas as pd
import numpy as np
import glob
```

```python
filenameArray = glob.glob('*.dat')
```

```python
# make list of input files with fake ab/c and phase values for now

filenameCol = filenameArray
typeCol = ['ab']*len(filenameArray)
phaseCol = np.random.rand(len(filenameArray)) # these are fake values

d = {'file': filenameCol, 'type': typeCol, 'phase': phaseCol}

input_list = pd.DataFrame(d, columns=['file','type','phase'])

input_list.to_csv('./rrlyrae_metallicity/src/spec_phases_fake.list', sep = ' ', header=False, index=False)
```

```python
# add errors (sqrt(flux)) to each spectral .dat file

for s in range(0,len(filenameArray)):
    filename = filenameArray[s]
    print(filename)
    data_array = pd.read_csv(filename, header=None, delim_whitespace=True)
    data_array[2] = np.sqrt(data_array[1])
    data_array.to_csv('./spectra_errors/'+filename, sep = ' ', header=False, index=False)
```
