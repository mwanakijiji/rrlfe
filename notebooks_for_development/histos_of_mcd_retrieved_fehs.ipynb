```python
import matplotlib.pyplot as plt
import pandas as pd
from pandas.plotting import scatter_matrix

%matplotlib qt
```

```python
plt.hist(df["feh_retrieved"], bins=100)
plt.title("McDonald spectra")
plt.xlabel("Retrieved [Fe/H]")
plt.show()
```
