```python
import glob
import os.path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# directory containing spectra in ascii format
source_dir = ("/Users/bandari/Documents/git.repos/rrlyrae_metallicity/"+
              "rrlyrae_metallicity/calib_application/src/original_backup_McD_spectra_apply_ascii/")

# directory to write new files with noise out to
write_dir = ("/Users/bandari/Documents/git.repos/rrlyrae_metallicity/"+
              "rrlyrae_metallicity/calib_application/src/McD_spectra_apply_ascii/")
```

```python
# accumulate list of filenames of *normalized* synthetic spectra

file_name_list = glob.glob(source_dir + "*smo") # (or search whatever other directory the *.c.dat files are in)

# for-loop to write out *.robolines and *.robospect files
for p in file_name_list: 
    print(p)
    df = pd.read_csv(p, delim_whitespace=True, names=["wavel","flux","error"])
    
    # make noise error equal to zero (or whatever)
    '''
    df["error"] = 1e-8
    '''
    df["error"] = np.divide(df["flux"],10)
    
    print(write_dir+os.path.basename(p))
    print('---')
    
    df.to_csv(write_dir+os.path.basename(p), sep=' ', index=False, header=False)
```

```python
# accumulate list of filenames of *UN-normalized* spectra

file_name_list = glob.glob(source_dir + "*_*dat") 

# for-loop to write out *.robolines and *.robospect files
for p in file_name_list: 
    print(p)
    df = pd.read_csv(p, delim_whitespace=True, names=["wavel","flux","error"], skiprows=1)
    
    # make noise error equal to zero (or whatever)
    '''
    df["error"] = 1e-8
    '''
    df["error"] = np.sqrt(df["flux"])
    
    print(write_dir+os.path.basename(p))
    print('---')
    
    df.to_csv(write_dir+os.path.basename(p), sep=' ', index=False, header=False)
```
