```python
import matplotlib.pyplot as plt
import pandas as pd
import glob
import os
```

```python
stem = "/Users/bandari/Documents/git.repos/rrlfe/rrlfe_io_20220803_01_mcd/realizations_output/norm/"
```

```python
file_list = glob.glob(stem + "*.dat")
```

```python
for file_name in file_list:
    
    df = pd.read_csv(file_name, delim_whitespace=True, names=["wavel","flux","continuum"])
    
    file_name_write = "plots/" + os.path.basename(file_name).split(".dat")[0] + ".png"
    
    plt.clf()
    plt.title(os.path.basename(file_name))
    plt.plot(df["wavel"], df["flux"])
    plt.plot(df["wavel"], df["continuum"])
    plt.savefig(file_name_write)
    print("Wrote", file_name_write)
```
