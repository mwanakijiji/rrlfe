```python
# This makes a plot of the X Ari spectrum around the Ca II K line, showing how
# small the ISM Ca II K line is

# Created 2019 June 15 by E.S.
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

### Read in spectroscopy of X Ari 
### (High-res is from the data used in Nemec, J., et al. 2013, ApJ, 773, 181)

```python
spec_highres = pd.read_csv("data/xari.spe", delim_whitespace = True)
spec_smoothed = pd.read_csv("data/xari.smo", delim_whitespace = True)
```

### Other line identities determined with SPECTRUM (see data/Metallines.dat)

### Make the plot

```python
fig = plt.figure(figsize = (8,6))
plt.rcParams.update({'font.size': 13})
subplot1 = fig.add_subplot(111)
subplot1.plot(spec_highres["wavel"], spec_highres["flux"], color = "k", label = "$R\sim36,000$")
subplot1.plot(spec_smoothed["wavel"], spec_smoothed["flux"], color = "r", label = "$R=2,000$")
subplot1.annotate("Ti II", (3913.460-1,0.3))
subplot1.annotate("Fe I", (3920.260-1,0.47))
subplot1.annotate("Fe I", (3922.910-1,0.4))
subplot1.annotate("Fe I", (3927.920-1,0.4))
subplot1.annotate("Fe I", (3930.300-1,0.4))
subplot1.annotate("Al I", (3944.010-1,0.58))
subplot1.annotate("Ca II K\n(ISM)", (3937-0.8,0.25))
subplot1.annotate("Ca II K\n(stellar)", (3930-0.5,0.1))
subplot1.arrow(3936, 0.3, -1.2, 0, length_includes_head = True, head_width = 0.03, head_length = 0.3, color = "k")
subplot1.set_xlim([3918,3950])
subplot1.legend()
subplot1.set_ylabel("Normalized Flux")
subplot1.set_xlabel("Wavelength ($\AA$)")
```

### Save the plot

```python
fig.savefig("x_ari_caiik_ism.pdf", bbox_inches='tight', pad_inches=0)
```
