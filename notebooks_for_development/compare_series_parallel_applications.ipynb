```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib qt
```

```python
#stem = '/Users/bandari/Documents/git.repos/rrlfe/rrlfe_io_20230318_quick_test/bin/'
stem = '/Users/bandari/Documents/git.repos/rrlfe/rrlfe_io_20221220_sdss_test/bin/'
```

```python
data_series = stem + 'retrieved_vals.csv'
data_parallel = stem + 'retrieved_vals_parallel_test.csv'
```

```python
df_series = pd.read_csv(data_series)
df_parallel = pd.read_csv(data_parallel)
```

```python
# merge

df_merged = df_series.merge(df_parallel, on="orig_spec_file_name", suffixes=('_series','_parallel'))
```

```python
plt.scatter(df_merged["feh_retrieved_series"], df_merged["feh_retrieved_parallel"])
plt.plot([-6000,9000],[-6000,9000], linestyle="--", color="k")
plt.xlabel("series")
plt.ylabel("parallel")
plt.legend()
plt.show()
```
