```python
# Takes SDSS spectra, and 
# 1. chops them to the range 3911 to 4950 A
# 2. uses blank space delimiters to make them ingestible into the pipeline

# Created 2022 Dec 20 by E.S.
```

```python
import pandas as pd
import glob
import numpy as np
import os
```

```python
stem = "/Users/bandari/Documents/git.repos/rrlfe/src/"
```

```python
file_list_read = glob.glob(stem + "sdss_20221213_cosmic_rays_removed_automated_full_wavel_range/" + "*csv")
```

```python
for file_num in range(0,len(file_list_read)):
    
    print(str(file_num) + "out of " + str(len(file_list_read)))
    
    df = pd.read_csv(file_list_read[file_num], names=["wavel","flux","error"])
    idx = np.logical_and(df["wavel"] >= 3911., (df["wavel"] <= 4950.))
    df[idx].to_csv(stem + "sdss_20221213_cosmic_rays_removed_automated_3911_to_4950_angstr/" + os.path.basename(file_list_read[file_num]), 
               sep=" ", index=False, header=False)
```
