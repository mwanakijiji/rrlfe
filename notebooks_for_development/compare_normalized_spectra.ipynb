```python
# This writes out images with normalized spectra which are overplotted if
# they have the same Teff and logg. This acts as a diagnostic of what error
# the [Fe/H]=-3.0 spectra might be introducing

# Created 2021 03 30 by E.S.
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import glob
```

```python
stem = "/Users/bandari/Documents/git.repos/rrlyrae_metallicity/rrlyrae_metallicity/realizations_output/norm/final/"
```

```python
# list of relevant file names

array_teff = ["5750","6000","6250","6500","6750","7000","7250","7500","7750"]
array_logg = ["20","25","30"]
array_feh = ["p02","p00","m05","m10","m15","m20","m25","m30"]

for i_t in range(0,len(array_teff)):
    for i_l in range(0,len(array_logg)):
        
        plt.clf()
        
        for i_f in range(0,len(array_feh)):
            
            plot_name = array_teff[i_t] + "_" + array_logg[i_l] + ".pdf"
            file_string = array_teff[i_t] + array_logg[i_l] + array_feh[i_f] + "_000"
            
            try:
                df = pd.read_csv(stem + file_string, delim_whitespace=True, names=["wavel","flux_norm"])
                plt.plot(np.add(i_f*0.,df["wavel"]),df["flux_norm"],label="[Fe/H] = " + array_feh[i_f], linewidth=0.5)
            except:
                print("Skipping " + file_string)
        
        plt.xlim([3800,5500])
        plt.title("Teff = " + array_teff[i_t] + ", logg = " + array_logg[i_l])
        plt.legend()
        plt.savefig(plot_name)
        print("Saved " + plot_name)
            

#file_list = glob.glob(stem + "*smo")
```
