```python
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt

%matplotlib qt
```

```python
#file_name = "/Users/bandari/Downloads/spec-0273-51957-0523.fits"
#file_name = "/Users/bandari/Downloads/spec-0273-51957-0523_full.fits"
file_name = "/Users/bandari/Downloads/spec-2867-54479-0613.fits"
```

```python
hdul = fits.open(file_name)
```

```python
data_blue = hdul[4].data
flux_blue = data_blue["flux"]
wavel_blue = np.power(10.,data_blue["loglam"])

data_red = hdul[20].data
flux_red = data_red["flux"]
wavel_red = np.power(10.,data_red["loglam"])
```

```python
plt.plot(wavel_red,flux_red)
plt.plot(wavel_blue,flux_blue)
plt.xlabel("wavel (angstrom)")
plt.ylabel("flux")
plt.show()
```
